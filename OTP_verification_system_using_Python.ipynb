# *OTP VERIFICATION USING PYTHON*

```python
import random
import smtplib
from tkinter import *
from tkinter.font import Font

attempts = 3  # Set the maximum number of attempts

def generate_otp():
    """Generate a random 6-digit OTP."""
    return random.randint(100000, 999999)

def send_otp():
    """Send OTP to the provided email."""
    global attempts
    receiver_email = email_entry.get()
    server = smtplib.SMTP('smtp.gmail.com', 587)
    server.starttls()
    server.login("shubhampatel91@gmail.com", "gmaaiwzmahfnddzo")
    
    otp = generate_otp()
    message = f"Subject: OTP verification\n\nDear {name_entry.get()},\n\nYour OTP is {otp}."
    server.sendmail("shubhampatel91@gmail.com", receiver_email, message)

    Label(window, text=f"OTP has been sent to {receiver_email}", bg="#ADD8E6", font=("Arial", 12)).grid(row=5, column=0, columnspan=2)

    verify_button = Button(window, text="Verify OTP", command=lambda: verify_otp(receiver_email, otp), bg="#4682B4", fg="white", font=("Arial", 12))
    verify_button.grid(row=6, column=0, columnspan=2)

def verify_otp(receiver_email, otp):
    """Verify the entered OTP."""
    global attempts
    entered_otp = otp_entry.get()
    if not entered_otp:
        Label(window, text="OTP not entered", fg="red", bg="#ADD8E6", font=("Arial", 12)).grid(row=7, column=0, columnspan=2)
        return

    # Remove the "OTP not entered" message if it exists
    for widget in window.grid_slaves():
        if widget.grid_info()["row"] == 7 and widget.grid_info()["column"] == 0:
            widget.grid_forget()

    entered_otp = int(entered_otp)
    if entered_otp == otp:
        bold_font = Font(weight="bold")
        Label(window, text="OTP verified", fg="green", bg="#ADD8E6", font=bold_font).grid(row=7, column=0, columnspan=2)
        # Remove the message of wrong attempt
        wrong_attempt_label.grid_forget()
    else:
        attempts -= 1
        if attempts > 0:
            # Display the "Invalid OTP" message in bold
            bold_font = Font(weight="bold")
            Label(window, text=f"Invalid OTP, {attempts} attempts left", fg="red", bg="#ADD8E6", font=bold_font).grid(row=7, column=0, columnspan=2)
        else:
            Label(window, text="You have exhausted all attempts", fg="red", bg="#ADD8E6", font=("Arial", 12)).grid(row=7, column=0, columnspan=2)

window = Tk()
window.title("OTP Verification")
window.geometry("400x200")
window.configure(bg="#ADD8E6")  # Set background color to light blue

Label(window, text="Enter your name:", bg="#ADD8E6", font=("Arial", 12)).grid(row=0, column=0)
name_entry = Entry(window, font=("Arial", 12))
name_entry.grid(row=0, column=1)

Label(window, text="Enter your email:", bg="#ADD8E6", font=("Arial", 12)).grid(row=1, column=0)
email_entry = Entry(window, font=("Arial", 12))
email_entry.grid(row=1, column=1)

Label(window, text="Enter OTP received:", bg="#ADD8E6", font=("Arial", 12)).grid(row=3, column=0)
otp_entry = Entry(window, font=("Arial", 12))
otp_entry.grid(row=3, column=1)

send_button = Button(window, text="Send OTP", command=send_otp, bg="#4682B4", fg="white", font=("Arial", 12))
send_button.grid(row=2, column=0, columnspan=2)

# Label to display wrong attempt message
wrong_attempt_label = Label(window, text="", fg="red", bg="#ADD8E6", font=("Arial", 12))
wrong_attempt_label.grid(row=7, column=0, columnspan=2)

window.mainloop()
```
